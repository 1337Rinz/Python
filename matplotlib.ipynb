```python
## import Library
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(10,25)
y = np.sin(x) # sin of x, it will convert the result of x to sin, after conversion we will get actual y
```

##Line plot


```python
## Line plot

plt.plot(x, y) # here plot is function, by default is line plt, we can specify other also
plt.show() # its recommended to show the visualization, show is kind of print function for visualization
```

```python
# Visualization
x = np.linspace(10, 25)
y = np.sin(x) * np.cos(x) ** 2
```

```python
plt.plot(x, y) # plot x over y; there are two axis (x & y)
plt.show()
```

```python
## Adding Label

# Label takes str inside function
plt.plot(x, y)
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.show()
```

```python
# Grid - default is false
plt.plot(x, y)
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.show()
```

```python
# Adding Figure Size
plt.figure(1, figsize=(20, 10)) # 1: Number of viz, 20: width, 10: height
plt.plot(x, y)
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.show()
```

```python
# Colour & Marker
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='lightgreen', marker = '1')
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.show()
```

```python
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='lightgreen', marker = '2',  linewidth=3)
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.show()
```

```python
plt.figure(1, figsize=(10, 8)) 
plt.plot(x, y, color ='red', marker = '*',  linewidth=3)
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.show()
```

```python
plt.figure(1, figsize=(10, 8)) 
plt.plot(x, y, color ='purple', marker = 'o',  linewidth=3) # marker is circle/ o
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.show()
```

```python
# Adding Title
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='purple', marker = 'o',  linewidth=3) # marker is circle/ o
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.title("Sample Visualization")
plt.show()
```

```python
#ticks
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='purple', marker = 'o',  linewidth=3) # marker is circle/ o
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.title("Sample Visualization")
plt.xticks(np.arange(10,25))  # ticks will control the x value range
plt.show()
```

```python
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='green', marker = 'o',  linewidth=3) # marker is circle/ o
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.title("Sample Visualization")
plt.xticks(np.arange(10,26,2))  # ticks with step size
plt.show()
```

```python
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='green', marker = 'o',  linewidth=3, label='Sin Wave') # marker is circle/ o
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.title("Sample Visualization")
plt.xticks(np.arange(10, 26, 2))  # ticks with step size
plt.yticks(np.arange(-.7, .7, 0.1))
plt.legend()
plt.show()
```

```python
# Labels & Legend

# Visualization 3- Legend with z
x = np.linspace(10, 25)
y = np.sin(x) * np.cos(x) ** 2
z = np.sin(x) ** 2  # square of sin wave
```

```python
plt.figure(1, figsize=(20, 10)) 
plt.plot(x, y, color ='green', marker = 'o',  linewidth=3, label='Sin Wave') # marker is circle/ o
plt.plot(x, z, color ='blue', marker = '^',  linewidth=3, label='Z Wave') 
plt.xlabel("Values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.title("Sample Visualization")
plt.xticks(np.arange(10, 26, 2))  # ticks with step size
plt.yticks(np.arange(-1, 1, 0.1))
plt.legend()
plt.show()
```

##Histogram
- Bins: Do value counts by range/ bins
- x axis: value range/ distribution
- y axis: Number  of counts withing range
- it will take a single param & divide into number of bins
- Example: Number of employees within salary range / distribution

```python
## Histogram
plt.figure(1, figsize=(20, 10))
plt.hist(y) # it will take a single param & divide into number of bins
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.grid(True)
plt.title("Range of value")
plt.show()
```

```python
plt.figure(1, figsize=(20, 10))
plt.hist(y, bins= 20, color='green') # Bin size & color customization)
plt.xlabel ("values")
plt.ylabel ('Sin(x)*cos(x)**2')
plt.grid(True)
plt.title('Sample visualization')
plt.show()
```

##Bar plot
- Deal with Categorical Distribution
- Histogram take value range & count the frequency within the range or distribution
- Barplot divide into categorical distribution.
- Barplot represent the count/ frequence withing a population or given dataset, whereas histogram show the frequency within give range/distribution.

```python
plt.figure(1, figsize=(20, 10))
plt.bar(['Python','Java','C++','JavaScript'],[9., 8., 8.5, 9.1])
plt.xlabel ('Values')
plt.ylabel ('Frequency')
plt.grid(True)
plt.title("Sample Visualization")
plt.show()
```

```python
plt.figure(1, figsize=(20, 10))
plt.barh(['Python','Java','C++','JavaScript'],[9., 8., 8.5, 9.1]) # barh:Horizontal Bar plot
plt.xlabel ('Values')
plt.ylabel ('Frequency')
plt.grid(True)
plt.title("Sample Visualization")
plt.show()
```

##Scatter Plot
- x & y values will be plotted as discrete data points
- Multiple plots can be generated at a time


```python
plt.figure(1, figsize=(16, 8))
x = np.random.random(200) # random value generation, x & y must be same in shape or size e.g 200
y = np.random.random(200)
plt.scatter(x, y)
plt.xlabel ('X')
plt.ylabel ('Y')
plt.grid(True)
plt.title('X vs Y')
plt.show()
```

```python
plt.figure(1, figsize=(16, 8))
x = np.random.random(200)
y = np.random.random(200)
plt.scatter(x, y, color='purple', marker='^') # Customization
plt.xlabel ('X')
plt.ylabel ('Y')
plt.grid(True)
plt.title("X vs Y")
plt.show()
```

```python
plt.figure(1, figsize=(16, 8))
x = np.random.random(200)
y = np.random.random(200)
plt.scatter(x, y, color='purple', marker='^') # Customization
plt.scatter(x, y*2, color='blue', marker='4') # Customization on y 2

plt.xlabel ('X')
plt.ylabel ('Y')
plt.grid(True)
plt.title("X vs Y")
plt.show()
```

```python
plt.figure(1, figsize=(16, 8))
x = np.random.random(200)
y = np.random.random(200)
plt.scatter(x, y, color='purple', marker='^') # Customization
plt.scatter(x*2, y, color='blue', marker='4') # More customization with x 

plt.xlabel ('X')
plt.ylabel ('Y')
plt.grid(True)
plt.title("X vs Y")
plt.show()
```

### xlim & ylim
- plt.xlim() : allow setting the limit for x axis, customization allow
- Customization size / can set limit from left & right
- it used then there are not data out of the range


```python
plt.figure(1, figsize=(16,8))
x = np.random.random(200)
y = np.random.random(200)
plt.scatter(x, y, color='purple', marker='o')
plt.scatter(x*2, y, color='blue', marker='^')
plt.xlabel ('X')
plt.ylabel ('Y')
plt.grid(True)
plt.title("X vs Y")
plt.legend("Test")
plt.ylim(.25, .8)
plt.xlim(.26, 1.75) # set limit: left / x will be started from .25 or range (.25 : -1.75)
plt.show()
```

```python
plt.figure(1, figsize=(16,8))
x = np.random.random(200)
y = np.random.random(200)
plt.scatter(x, y, color='purple', marker='o')
plt.plot(x*2, y, color='blue', marker='4') # Line plot,  multiple plot at a time
plt.xlabel ('X')
plt.ylabel ('Y')
plt.grid(True)
plt.title("X vs Y")
plt.show()
```

##3D Visualization

```python
import random

# Create an empty array with 50 elements
x = [0.0 for i in range(50)]

# Populate the array with random values between 0.0 and 1.0
for i in range(50):
    x[i] = random.uniform(0.0, 1.0)

print(x)
```

```python
# Create an empty array with 50 elements
y = [0.0 for i in range(50)]

# Populate the array with random values between 0.0 and 1.0
for i in range(50):
    y[i] = random.uniform(0.0, 1.0)

print(y)
```

```python
# Create an empty array with 50 elements
z = [0.0 for i in range(50)]

# Populate the array with random values between 0.0 and 1.0
for i in range(50):
    z[i] = random.uniform(1.0, 10.0)

print(z)
```

```python
x = [float(round(x,2)) for x in x]
y = [float(round(y,2)) for y in y]
```

- need 3 values (e.g x, y, z)

```python
axes = plt.axes(projection='3d') # This is pre-requisition for 3d visualization
```

```python
axes = plt.axes(projection='3d')
z = [x * y for x, y in zip(x, y)]
axes.scatter3D(x, y, z)
plt.show()
```

###3D plot Customization
- in 3D plot interaction with other plot not possible
- Jupyter notebook in brower & pyfile will allow the interactivity
- For interaction Java Script used in backend

```python
# Adding Labels
axes = plt.axes(projection='3d')
z = [x * y for x, y in zip(x, y)]
axes.scatter3D(x, y, z)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.show()
```

```python
# color Customization
axes = plt.axes(projection='3d')
z = [x * y for x, y in zip(x, y)]
axes.scatter3D(x, y, z, color ='red')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
```

```python
# Marker Customization

axes = plt.axes(projection='3d')
z = [x * y for x, y in zip(x, y)]
axes.scatter3D(x, y, z, color ='purple', marker='^')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
```

```python
# Size Customization
# color Customization
plt.figure(1, figsize=(16,8))
axes = plt.axes(projection='3d')
z = [x * y for x, y in zip(x, y)]
axes.scatter3D(x, y, z, color ='green', marker ='^')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
```
